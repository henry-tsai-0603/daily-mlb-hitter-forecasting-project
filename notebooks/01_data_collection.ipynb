Collect Statcast data using pybaseball and save raw files.



```python
import pandas as pd
# import numpy as np
import sys
from pathlib import Path

# Make sure notebook can import src/
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
from pybaseball import statcast

df = statcast(start_dt="2025-04-01", end_dt="2025-04-03")
df.shape
```

```python
df.columns.tolist()
```

```python
needed_cols = [
    "game_date",
    "batter",
    "pitcher",
    "player_name",
    "events",
    "description",
    "pitch_type",
    "stand",
    "p_throws",
    "launch_speed",
    "launch_angle",
    "estimated_woba_using_speedangle",
    "woba_value",
]

missing_cols = [c for c in needed_cols if c not in df.columns]

print("Missing columns:", missing_cols)
```

```python
df_sample = df[needed_cols].copy()

print(df_sample.shape)
df_sample.head()
```

```python
df_sample.info()
df_sample.isna().mean().sort_values(ascending=False)
```

```python
output_path = PROJECT_ROOT / "data/raw/statcast_sample_2025_04_01_to_04_03.csv"

save_csv(df_sample, output_path)

print(output_path)
print("File exists:", output_path.exists())
```

```python
pa_df = df_sample[df_sample["events"].notna()].copy()

print(pa_df.shape)
pa_df.head()
```

```python
pa_df["events"].value_counts().head(30)
```

```python
pa_df[["woba_value", "estimated_woba_using_speedangle"]].isna().mean()
```

```python
pa_df["xwoba_value"] = pa_df["estimated_woba_using_speedangle"].fillna(pa_df["woba_value"])

pa_df[["events", "woba_value", "estimated_woba_using_speedangle", "xwoba_value"]].head(20)
```

```python
pa_df["xwoba_value"].isna().mean()
```

```python
hitter_summary = (
    pa_df
    .groupby(["batter", "player_name"])
    .agg(
        PA=("events", "count"),
        xwOBA=("xwoba_value", "mean"),
        wOBA=("woba_value", "mean"),
        avg_exit_velocity=("launch_speed", "mean"),
        avg_launch_angle=("launch_angle", "mean"),
    )
    .reset_index()
    .sort_values("PA", ascending=False)
)

hitter_summary.head(20)
```

```python
processed_path = PROJECT_ROOT / "data/processed/pa_sample_2025_04_01_to_04_03.csv"

save_csv(pa_df, processed_path)

print(processed_path)
print("File exists:", processed_path.exists())
```

```python
summary_path = PROJECT_ROOT / "data/processed/hitter_summary_sample_2025_04_01_to_04_03.csv"

save_csv(hitter_summary, summary_path)

print(summary_path)
print("File exists:", summary_path.exists())
```
